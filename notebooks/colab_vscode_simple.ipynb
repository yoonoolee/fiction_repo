# Connect VS Code to Colab (Simple Method)

Uses Cloudflare tunnel (no signup needed!)

**Steps:**
1. Run Cell 1 (install)
2. Run Cell 2 (start SSH)
3. Copy SSH command from output
4. In VS Code: Cmd+Shift+P → Remote-SSH: Connect to Host
5. Password: `colab`
6. Run Cell 3 to keep session alive

```python
# Cell 1: Install dependencies
!pip install colab-ssh --upgrade -q
print(" Installation complete!")
```

```python
# Cell 2: Start SSH server with Cloudflare
from colab_ssh import launch_ssh_cloudflared

print(" Starting SSH server...")
print("⏳ This takes ~30 seconds...\n")

launch_ssh_cloudflared(password="colab")

print("\n" + "="*60)
print(" SSH SERVER IS RUNNING!")
print("="*60)
print("\n COPY the 'ssh root@...' command above")
print(" Password: colab")
print("\n In VS Code: Cmd+Shift+P → Remote-SSH: Connect to Host")
print("="*60)
```

```python
# Cell 3: Keep session alive (run this after connecting VS Code)
import time
from IPython.display import clear_output

print("⏰ Keep-alive loop running...")
print(" Don't stop this cell or your connection will drop!\n")

while True:
    clear_output(wait=True)
    print(f" Session active: {time.strftime('%H:%M:%S')}")
    print(" VS Code connected via SSH")
    print("  Don't stop this cell!")
    time.sleep(60)
```
